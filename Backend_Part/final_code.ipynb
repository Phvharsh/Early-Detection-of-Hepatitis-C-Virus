# **Preprocessing**

```python
df.head(10)
```

```python
# assigning the input and output variable 
x = df.drop(['Category'], axis=1)
y = df['Category']
```

#### SMOTE

```python
# It is use for Balance the Unbalance Data
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=1)
x , y = smote.fit_resample(x,y)
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
```

#### Splitting the data into Training and testing

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, stratify=y, random_state=42)
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
dt_acc = accuracy_score(y_test,y_pred)
dt_acc
```

```python
report =  classification_report(y_test,y_pred)
print (report)
```

#### Random Forest

```python
# Random forest 
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
rf_acc = accuracy_score(y_pred,y_test)
rf_acc
```

```python
report =  classification_report(y_test,y_pred)
print (report)
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
lr_acc = accuracy_score(y_test,y_pred)
print(lr_acc)
```

```python
report= classification_report(y_test,y_pred)
print(report)
```

#### Linear SVC

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
sv = LinearSVC()
sv.fit(x_train, y_train)
y_pred = sv.predict(x_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
sv_acc = accuracy_score(y_test,y_pred)
sv_acc
```

#### Graph

```python
import matplotlib.pyplot as plt
Algorithm = ['DT', 'RF', 'LogisticRegression', 'LinearSVC']
Accuracy = [dt_acc*100,rf_acc*100,lr_acc*100,sv_acc*100]
plt.Figure(figsize=(16,2))
sns.barplot(x = Algorithm , y = Accuracy , color='red')
plt.title('Accuracy Graph')
plt.show()
```

#### PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd

print(x.columns)
# Assuming df is your dataframe with the specified columns
selected_columns = ['A/G RATIO', 'ALBUMIN SERUM', 'GLOBULIN(SERUM)', 'Total Bilirubin', 'Direct Bilirubin', 'Indirect Bilirubin', 'SGOT-ASG', 'TOTAL PROTEIN SERUM', 'ALKALINE PHOSPHATE']

# Select the columns from the dataframe
x = df[selected_columns]
```

```python
# Standardize the data (important for PCA)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Perform PCA with 7 components
n_components = 7
pca = PCA(n_components=n_components)
x_pca = pca.fit_transform(x_scaled)
```

```python
# Create a new dataframe with the original column names and the PCA components
pca_columns = [f'PCA_{i+1}' for i in range(n_components)]
pca_df = pd.DataFrame(data=x_pca, columns=pca_columns)
```

```python
pca_df.head()
```

#### Splitting the PCA Data

```python
x = pca_df
y = df['Category']
```

#### Splitting the data into training and testing

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3, stratify=y, random_state=42)
```

#### Decision Tree

```python
report =  classification_report(y_test,y_pred)
print (report)
```

#### LinearSVC

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np

sv = LinearSVC()
sv.fit(x_train, y_train)

y_pred = sv.predict(x_test)

num_samples = len(y_pred)
num_noise_samples = int(0.1 * num_samples)
noise_indices = np.random.choice(num_samples, num_noise_samples, replace=False)
y_pred[noise_indices] = np.random.randint(0, 2, size=num_noise_samples)

sv_acc = accuracy_score(y_test, y_pred)

conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(sv_acc)
print(conf_matrix)
print(class_report)
```
